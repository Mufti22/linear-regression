```python
def PLOT34(y_test, y_pred):
    '''
    Функция построения графиков
    :param y_test: - проверочные значения целевой переменный
    :param y_pred: - вычисленные значения целевой переменный
    '''
    plt.figure(figsize=(12,6))
    
    # == Диаграмма рассеяния вычисленных значений ==
    # создать первое полотно 121: 1-строк, 2-столбцов, 1-индекс текущего полотна в сетке
    plt.subplot(121)
    plt.scatter(y_test, y_pred,  alpha=0.1)
    plt.scatter(y_test, y_test,  alpha=0.1)
    plt.title('Диаграмма рассеяния вычисленных значений');
    plt.xlabel('Проверочное Y')
    plt.ylabel('Вычисленное Y')

    
    # == Диаграмма рассеяния ошибок ==
    # создать первое полотно 121: 1-строк, 2-столбцов, 2-индекс текущего полотна в сетке
    plt.subplot(122)
    plt.scatter(y_test, (y_test - y_pred)**2,  alpha=0.1)
    plt.title('Диаграмма рассеяния квадрата абсолютной ошибки')
    plt.xlabel('Проверочное Y')
    plt.ylabel('Квадрат абсолютной ошибки')
#--------------------------------------------------------------------------
```

```python
def GET_METRICS_SINGLE(y_test, y_pred):
    '''
    Вычисление и вывод метрик: MAE, RMSE, R2.
    На основе сравнения проверочных и вычисленных.
    :param y_test: - проверочные значения целевой переменный
    :param y_pred: - вычисленные значения целевой переменный
    '''
    mae  = metrics.mean_absolute_error        (y_test, y_pred)
    mse  = metrics.mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2   = metrics.r2_score                   (y_test, y_pred)

    print ("MAE : {:>9,.3f} (средняя абсолютная ошибка)".format              ( mae ))
    print ("MSE : {:>9,.6f} (среднеквадратичная ошибка)".format( mse ))
    print ("RMSE: {:>9,.6f} (кв. корень из среднеквадратичной ошибки)".format( rmse ))
    print ("R2  : {:>9,.3f} (коэфф. детерминации)".format                    ( r2 ))
#--------------------------------------------------------------------------
```

```python
df = pd.read_csv('./diamonds_true_prepared.csv', sep = ',', header = 0)
print(df.head())
```

```python
features = ["carat","x","y","z","clarity"]
target = ["price"]
```

```python
dfX = df[features]
print(dfX.head())
dfY = df[target]
print(dfY.head())
```

```python
print(min(df['price']))
```

```python
NormdfX = pd.DataFrame(data = MinMaxScaler().fit(dfX).transform(dfX), columns = features)
print(NormdfX.head())
NormdfY = pd.DataFrame(data = MinMaxScaler().fit(dfY).transform(dfY), columns = target)
print(NormdfY.head())
```

Не нормализованные исходные данные

```python
test_size = 0.4
rng = 42
```

```python
valid_size = 0.3 # доля тестовой части в выборке
rand_seed = 8 # начальное состояние генератора случ. чисел
x_train, x_test, y_train, y_test = train_test_split(
        dfX,
        dfY,
        test_size = test_size,
        random_state = rng,
        shuffle = True)
```

Линейная регрессия для исходных данных

```python
lr = linear_model.LinearRegression()

#Обучаем
linear_model_fitted = lr.fit(x_train, y_train)
y_lin_pred = linear_model_fitted.predict(x_test)
```

```python
GET_METRICS_SINGLE(y_test, y_lin_pred)
PLOT34(y_test, y_lin_pred)
```

На нормализированных данных. Нейронная сеть со скрытыми слоями, сделанная по уму(всё-таки случайно, но преподаватель сказал что будет лучше так, чтобы внутренние слои постепенно сходили от многих нейронов в скрытых слоях до одного на выходном слое)

С приведённой ниже структурой были проведены тесты с разными данными(я решил сюда их не включать, так как они считаются долго, а их результаты неприемлимы вплоть до того, что коэффициент детерминации был равен +-0.4, а показатели ошибки были вплоть до 0.3)
Сначала использовалось больше слоёв с большим количеством скрытых нейронов(8 слоёв, количество нейронов 150). Такая ситуация только на первых 10 эпохах выдавала параметры сравнимые с показанными с используемыми характеристиками.

```python
with tf.device('/CPU:0'):
    totalLossTrain = []
    totalLossTest = []
    
    
    #Параметры сети
    input_size = 5
    output_size = 1
    
    
    #Начинаем строить сеть
    model = tf.keras.models.Sequential() #Объект класса секуентиал. Стек слоёв.    
    
    model.add(tf.keras.layers.Input(shape = (input_size,)))
    model.add(tf.keras.layers.Dense(units = 38, activation = tf.keras.activations.tanh))
    model.add(tf.keras.layers.Dense(units = 18, activation = tf.keras.activations.tanh))
    model.add(tf.keras.layers.Dense(units = 10, activation = tf.keras.activations.tanh))
    model.add(tf.keras.layers.Dense(units = 5, activation = tf.keras.activations.tanh))
    model.add(tf.keras.layers.Dense(units = output_size, activation = tf.keras.activations.linear))
    
    LossFunction = tf.keras.losses.mean_squared_error
    OptimisationFunction = tf.keras.optimizers.Adam(learning_rate = 0.01)
    MetricFunction = [tf.keras.losses.mean_squared_error]
    
    model.compile(loss = LossFunction, optimizer = OptimisationFunction, metrics = [MetricFunction])
    
    print(model.summary())
```

```python
epochForTrain = 200

with tf.device('/CPU:0'):
    history = model.fit(
    
        Normx_train [['carat','x','y','z','clarity',]],
        Normy_train,
    
        validation_data = (
        Normx_test[['carat','x','y','z','clarity',]],
        Normy_test),
        
        epochs = epochForTrain,
        verbose = 2,)
    
    totalLossTrain.extend(history.history['loss'])
    if 'val_loss' in history.history.keys():
        totalLossTest.extend(history.history['val_loss'])
```

```python
totalLossTrain[-5:]
```

```python
plt.plot(totalLossTrain, label = 'train')
if 'val_loss' in history.history.keys():
    plt.plot(totalLossTest, label='test')
plt.legend()
plt.title("История изменений выбраной оценки потерь (LossVal = mean_squared_error)")
plt.ylabel("Оценка потерь (mean_squared_error)")
plt.xlabel("Эпохи обучения")
```

```python
# create model
with tf.device('/CPU:0'):
    # Calculate predictions
    Normy_pred = model.predict( Normx_test [['carat','x','y','z','clarity',]] ) # требуется передать таблицу с кол-ом столбцов равным кол-ву входных нейронов

GET_METRICS_SINGLE(Normy_test, Normy_pred)
PLOT34(Normy_test, Normy_pred)
```

```python
m1dict  = dict()
m1dict["ТипМодели"] = str(type(lr))
m1dict["Нормализация"] = False
m1dict["features"] = features
m1dict["target"] = target
m1dict["R2"]   = metrics.r2_score(y_test, y_lin_pred)
m1dict["RMSE"]   = np.sqrt( metrics.mean_squared_error(y_test, y_lin_pred))

m2dict  = dict()
m2dict["ТипМодели"] = str(type(model))
m2dict["Нормализация"] = True
m2dict["features"] = features
m2dict["target"] = target
m2dict["R2"]   = metrics.r2_score(Normy_test, Normy_pred)
m2dict["RMSE"]   = np.sqrt(metrics.mean_squared_error(Normy_test, Normy_pred))
```

```python
with open('Linear_Regression_model.json', 'w', encoding='utf-8') as filestream:
    json.dump(m1dict, filestream,  indent=2)
    print(f'Модель {type(m1dict)} была записана в файл Linear_Regression_model.json')
    
with open('Neural_Network.json', 'w', encoding='utf-8') as filestream:
    json.dump(m2dict, filestream,  indent=2)
    print(f'Модель {type(m2dict)} была записана в файл Neural_Network.json')
```
